# **Dias cronologicos al interior 2024 (enero-abril)**

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import (pairwise_tukeyhsd, MultiComparison)
import scipy.stats as stats
```

```
# Carga de dataset
df = pd.read_csv("https://raw.githubusercontent.com/edisonelcp1/HEALTH/main/diasinterior2024.txt", sep="\t")
df.head()
```

```
df.describe().T # Media, Desvesta, Percentiles
```

```
# Concatenar la fuente de la factura y el numero de factura en una nueva columna (al final queda agregada), se genera columna tipo fecha (de generación de la factura), el año y mes
df['fuentefact'] = df['Fuente'] + '-' + df['Nro Factura'].astype(str)
df['Date'] = pd.to_datetime(df["fecha_fact"])
df["mes"] = df["Date"].dt.month
df["ano"] = df["Date"].dt.year.astype(str)
df.head()
```

```
# Se eliminan los estados de la factura que no interesan y se eliminan las fuentes de factura FI, FM, y FP
df2=df.drop(['afga','aug','fag','cia','fae','efp','frp','rips','aief','aiem','aien','afee','ar01','ar02','ar03','ar04','ar05','ar06','ae01','ae02','am01','am02','am03','am04','fauf','aiep','asfp','afd','afn','afmm','afa','adf','afc','ar07','aftp','adpi','fcj','refac','fcp','aied','furips','qcd','penradi','nomina','fecha_fact'], axis=1)
df2.drop(df2[(df2['Fuente'] == 'FI') | (df2['Fuente'] == 'FM') | (df2['Fuente'] == 'FP')].index, inplace=True)
df2.head()
```

```
# Se valida que solo quedan los estados necesarios y las fuentes de factura FH y RF. Quedan 99589 facturas FH y 1731 refacturas RF para un total de 101320 facturas
ft = df2.groupby('Fuente').count()
ft.head()
```

```
ft.describe().T
```

```
# se crea función para percentiles 25, 75 y mediana
def P25(x):
    return x.quantile(0.25)

def med(x):
    return x.quantile(0.5)

def P75(x):
    return x.quantile(0.75)

# Las funciones creadas se aplicarán a la columna 'total_dias'
vals = {'total_dias': [P25, med, P75, 'count']}
```

```
grupo = df2.groupby(['Empresa','dependencia']).agg(vals).reset_index()
grupo.head()
```

```
grupo = df2.groupby(['Empresa']).agg(vals).reset_index()
grupo.head()
```
